```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
```

```python
model = ['Lasso Regression', 'Ridge Regression', 'Bayesian Ridge', 'non-negative Matrix Factorization', 'Singular Value Decomposition']
mae = [1.3457, 1.5025, 1.3138, 1.2136, 1.2413]
rmse = [2.0493, 2.3622, 2.0073, 1.8807, 1.8755]
ndcg = [0.9374, 0.9398, 0.9418, 0.9530, 0.9559]
x = np.arange(len(model))  # the label locations
width = 0.25  # the width of the bars


fig, ax = plt.subplots(figsize=(15,10))
rects1 = ax.bar(x - width, mae, width, label='MAE')
rects2 = ax.bar(x, rmse, width, label='RMSE')
rects3 = ax.bar(x + width, ndcg, width, label='nDCG')
font = {'family': 'serif',
        'color':  'darkred',
        'weight': 'normal',
        'size': 15,
        }
font1 = {'family': 'serif',
        'color':  'darkred',
        'weight': 'normal',
        'size': 20,
        }
font2 = {'family': 'san-serif',
         'color':  'black',
         'weight': 'normal',
         'size': 12,
         }
def autolabel(rects):
    """
    Attach a text label above each bar displaying its height
    """
    for rect in rects:
        height = rect.get_height()
        ax.text(rect.get_x() + rect.get_width()/2., 0.01 + height,
                '%1.4f' % height,
                ha='center', va='bottom', fontdict=font2)
autolabel(rects1)
autolabel(rects2)
autolabel(rects3)
# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Value', fontdict=font1)
ax.set_title('Evaluation', fontdict=font1)
ax.set_xticks(x)
ax.set_xticklabels(model, fontdict=font)
ax.legend()
plt.xticks(rotation=10)
fig.tight_layout()
plt.show()
```
